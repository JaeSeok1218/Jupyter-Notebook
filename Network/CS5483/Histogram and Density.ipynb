```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Histogram Comparison

```python
G1 = nx.barabasi_albert_graph(100,1) #A random graph skewed with a few central nodes(1 central nodes)
print([G1.number_of_nodes(), G1.number_of_edges()])
nx.draw_spring(G1)
```

```python
plt.hist([v for k,v in nx.degree(G1)])
plt.title('Histogram')
plt.show()
```

```python
degree_sequence = [d for n, d in G1.degree()]
degreeHist = {} # create a vector (or think of this as an array)
for d in degree_sequence:
    if d in degreeHist:
        degreeHist[d]+= 1 #if there is a degree d increment the number of occurrances by one
    else:
        degreeHist[d] = 1
print ("degree #of_nodes")
for d in degreeHist:
    print(f"{d:4}  {degreeHist[d]:6}") # the :4 or :6 is number of spaces for that field when printing
```

```python
G2 = nx.barabasi_albert_graph(100,4) #A random graph skewed with a few central nodes(4 central nodes)
print([G2.number_of_nodes(), G2.number_of_edges()])
nx.draw_spring(G2)
```

```python
plt.hist([v for k,v in nx.degree(G2)])
plt.title('Histogram')
plt.show()
```

## Density and Mean degree of undirected graph

```python
density = nx.density(G1)
print("Network density", density)
nx.draw_networkx(G1, node_size=5, with_labels=False)
```

```python
m = sum(G1.degree(n) for n in G1.nodes())
n = G1.number_of_nodes()
print(f'm = {m},n = {n}')
print(2*m/n)
```

## Mean degree of directed graph

```python
DG1 = nx.gnp_random_graph(100,0.05,directed = True)
mOut = sum(DG1.out_degree(n) for n in DG1.nodes())
mIn = sum(DG1.in_degree(n) for n in DG1.nodes())
n = DG1.number_of_nodes()
print(f'm = {mOut} = {mIn},n = {n}')
print(mOut/n)
print(mIn/n)
```
