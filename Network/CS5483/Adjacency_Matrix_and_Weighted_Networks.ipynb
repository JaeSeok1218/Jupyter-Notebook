<h1 style="font-size:5rem;color:green;"> Adjacency Matrix and Weighted Networks </h1>

```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

UG = nx.fast_gnp_random_graph(10,0.2)
DG = nx.fast_gnp_random_graph(10,0.2, directed=True)
```

# Adjacency matrix, Edge list, and Adjacency list

```python
print ('Adjacency Matrix')
A = nx.adjacency_matrix(UG)
print(A.todense())
# Print the edge list
print ('Edge List')
for line in nx.generate_edgelist(UG, data=False):
    print(line)

nx.draw(UG, with_labels=True)
```

```python
print('Adjacency Marix')
B = nx.adjacency_matrix(DG)
print(B.todense())
# Print the edge list
print ('Edge List')
for line in nx.generate_edgelist(DG, data=False):
    print(line)

nx.draw(DG, with_labels=True)
```

```python
for line in nx.generate_adjlist(UG):
    print(line)
```

```python
for line in nx.generate_adjlist(DG):
    print(line)
```

# Weighted Networks

```python
V1 = ['A','B','C','D','E','F']
E1 = [('A','B',6),('B','C',7),('C','D',8),('D','E',4),('E','F',9),('F','A',6)]
G = nx.DiGraph()
G.add_nodes_from(V1)
G.add_weighted_edges_from(E1)
```

```python
pos = {'A':[1,1],'B':[3,1],'C':[4,2],'D':[3,3],'E':[1,3],'F':[0,2]}
```

```python
weight = nx.get_edge_attributes(G, 'weight')

nx.draw(G, pos=pos, with_labels=True, arrowsize=35)
nx.draw_networkx_edge_labels(G, pos, edge_labels = weight)
```

```python
V2 = ['A','B','C','D','E']
E2 = [('A','B',13),('A','C',10),('A','D',12),('B','D',7),('B','E',10),('C','D',9),('D','E',8)]
G1 = nx.Graph()
G1.add_nodes_from(V2)
G1.add_weighted_edges_from(E2)
```

```python
pos = nx.spring_layout(G1) # to draw edge_labels: positioning
weight = nx.get_edge_attributes(G1, 'weight')
nx.draw(G1, pos,
       with_labels=True)

edge_labels = dict([((n1, n2), d['weight'])
                    for n1, n2, d in G1.edges(data=True)]) # make a dictionary; n1-node n2-node d-'relation' here
nx.draw_networkx_edge_labels(G1,pos, edge_labels)
plt.show(G1)
```

```python
list1=[]
for (n1, n2, d) in G1.edges(data=True):
    d.clear()
    list1.append((n1,n2))
```

```python
G2 = nx.Graph()
G2.add_edges_from(list1)
nx.draw(G2)
```
