```python
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Contruct a graph

V = ['a','b','c','d','e','f','g','h','i']

G = nx.Graph()
G.add_nodes_from(V)
G.add_edge('a','b', relation='s')
G.add_edge('a','c', relation='w')
G.add_edge('a','d', relation='s')
G.add_edge('a','e', relation='w')
G.add_edge('a','f', relation='s')
G.add_edge('b','g', relation='w')
G.add_edge('b','i', relation='s')
G.add_edge('e','i', relation='w')
G.add_edge('f','h', relation='s')
G.edges(data=True)
```

```python
pos = nx.spring_layout(G) # to draw edge_labels: positioning
nx.draw(G, pos,
       node_color='C1',
       node_size=200,
       with_labels=True)

edge_labels = dict([((n1, n2), d['relation'])
                    for n1, n2, d in G.edges(data=True)]) # make a dictionary; n1-node n2-node d-'relation' here
nx.draw_networkx_edge_labels(G,pos, edge_labels)
plt.show(G)
```

## a. Degree Centrality

```python
d = nx.degree_centrality(G)

for v in G.nodes():
    print(f"{v:2} {d[v]:.3f}")
    
sort_list = sorted(d.items(), key=lambda x: x[1], reverse=True) 
# key=lambda x:x[1]: lambda function(simple function) sorting by 1st element

color_map = []
for node in G:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(G, node_color=color_map, with_labels=True)
```

## b. Eigenvector Centrality

```python
centrality = nx.eigenvector_centrality(G)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in G:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(G, node_color=color_map, with_labels=True)
```

## c. Page Rank Centrality

```python
centrality = nx.pagerank(G, alpha=0.5, personalization=None, max_iter=100, tol=1e-06, nstart=None, weight='weight', dangling=None)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in G:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(G, node_color=color_map, with_labels=True)
```

```python
centrality = nx.katz_centrality_numpy(G, alpha=0.2, beta=1.0, normalized=True, weight=None)
for v, c in centrality.items():
    print(f"{v:2} {c:.3f}")

sort_list = sorted(centrality.items(), key=lambda x: x[1], reverse=True)

color_map = []
for node in G:
    if (node == sort_list[0][0]) or (node== sort_list[1][0]):
        color_map.append('red')
    else: 
        color_map.append('green')      
nx.draw(G, node_color=color_map, with_labels=True)
```
