```python
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx
```

```python
el = open ("Module-1-EdgeList-1.txt", 'rb') # open file and rb is for read-binary
G = nx.read_edgelist(el, nodetype=int) # edges on each line of the input
el.close() # we can close the file now
```

```python
# This is a sequence generator. Note the square brackets the RHS of assignment statement.
# for n means for all vertices from 1 to n, 
degree_sequence = [d for n, d in G.degree()] 
```

```python
nx.draw_circular(G, node_color='C0',node_size=200,with_labels=True)
plt.show()
```

```python
G.degree () # it will give a you list of nodes and their degrees
```

```python
# for n is like a for-loop goes through each element in G.degree. d is the degree of each node in G.degree() list
degree_sequence = [d for n, d in G.degree()]  # degree sequence
```

```python
#node the f infront of ". It says evaluate, i.e. get the values in degree_sequence
print(f"Degree sequence {degree_sequence}") 
```

```python
# We now going to determine a histogram of degrees
# Python is all tab-oriented; Use Tabs to indent

degreeHist = {} # create a vector (or think of this as an array)
for d in degree_sequence:
    if d in degreeHist:
        degreeHist[d]+= 1 #if there is a degree d increment the number of occurrances by one
    else:
        degreeHist[d] = 1
print ("degree #of_nodes")
for d in degreeHist:
    print(f"{d:4}  {degreeHist[d]:6}") # the :4 or :6 is number of spaces for that field when printing
    
```
